```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

**En esta visualización veremos la relevancia de Lionel Messi para el FC Barcelona a lo largo de los años**

La primera visualización que veremos sera la influencia de Lionel Messi en sus años de estancia en el FC Barcelona.

*Contexto*:

Messi debuta en el primer equipo del FC Barcelona en el año 2004, en la temporada 2004/2005, desde ahí jugó 16 años seguidos hasta el año 2021, siendo la temporada 2021/2022 la primera temporada de el FC Barcelona sin Lionel. 

La primera visualización usará las estadísticas del FC Barcelona desde la temporada 2000/2001 hasta la última temporada 2024/2025. El proposito principal es ver como cambiaron las estadisticas generales del FC Barcelona a lo largo de las temporadas y visualizar el impacto de Lionel Messi en el equipo.

```python
# Leemos el dataset del Barça
df_barca = pd.read_excel("FC_Barcelona_Dataset_v2.xlsx")
df_barca["Equipo"] = "FC Barcelona"

# Leemos todas las pestañas de los demás clubes
xls = pd.ExcelFile("Grandes_Clubes_Dataset.xlsx")
sheet_keys = xls.sheet_names[0:-1] # excluyes la última pestaña

dfs = [df_barca]  # empezamos con el Barça

for sheet in sheet_keys:
    df_club = pd.read_excel(xls, sheet_name=sheet)
      # el nombre de la pestaña como nombre del equipo
    dfs.append(df_club)

# Combinamos todo en un solo DataFrame
df = pd.concat(dfs, ignore_index=True)
df["PG"] = df["PG"].fillna(0)

```

```python
fig = px.scatter(
    df,
    x="GF",
    y="PTS",
    size="PG",
    animation_frame="Temporada",
    color="Equipo",
    size_max=40,
    range_x=[20, 130],
    range_y=[20, 122],
    text="Equipo",
    color_discrete_map={
        "FC Barcelona": "#C70656",
        "Real madrid": "#FEBE10",
        "Atletico de Madrid": "#EB8F8F",
        "Liverpool FC": "#06D85A",
        "Manchester united": "#FF0000",
        "Villarreal CF": "#EAFF00",
        "Manchester City": "#31C8FF",
        "Bayern Munich": "#A83C3C",
        "Juventus": "#5B5959",
        "Inter de milan": "#050C6D",
        "AC Milan": "#500505",
        "Valencia CF": "#E17101",
        "Sevilla CF": "#E4C5C5",
        "Arsenal": "#FE00CF",
    },
    opacity=1
)

fig.update_layout(
    title={
        "text": "Equipos importantes de Europa en lo que va de milenio",
        "x": 1
    },
    legend=dict(
        title="Haz click en un equipo →",
        itemclick="toggle",
        itemdoubleclick="toggleothers"
    )
)

fig.update_traces(
    textposition="middle center",
    textfont=dict(size=9),
    marker=dict(line=dict(width=1, color="white"))
)

fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1200

fig.show()

"""El gráfico es interactivo, la idea es ir comparando el rendimiento del Barcelona, clickando los equipos en la leyenda lo desactivas"""
```

Viendo esta visualización podemos notar algunas cosas:
- Desde que Messi entró al equipo (2004/2005), solo 2 temporadas el Barcelona metió menos de 80 goles.
- La primera temporada sin Messi (2021/2022) el Barcelona metió menos de 80 goles por primera vez en 13 años.
- También podemos notar que las temporadas sin Messi fueron las con menos puntos de los últimos 10 temporadas atras.

Con esta visualización podemos notar el impacto de Messi en el equipo en el ámbito goleador, pesé a que los puntos no siempre fueron en alza, la faceta goleadora del argentino siempre se noto en el FC Barcelona, y por eso cuando se va del club los goles a favor fueron a la baja.
Podemos concluir que Messi es fundamental en la historia moderna del barcelona

```python
import pandas as pd
import plotly.express as px

df_messi = pd.read_excel("messi_goles_laliga.xlsx")

df_ciudad = df_messi.groupby(["Ciudad", "Latitud", "Longitud"])["Goles"].sum().reset_index()

fig = px.scatter_mapbox(
    df_ciudad,
    lat="Latitud",
    lon="Longitud",
    size="Goles",
    text="Ciudad",
    hover_name="Ciudad",
    hover_data={"Goles": True, "Latitud": False, "Longitud": False},
    size_max=40,
    color="Goles",
    color_continuous_scale=[
    [0, "#D3D3D3"],    # gris claro para valores bajos
    [0.5, "#FF6B6B"],  # rojo medio
    [1, "#8B0000"]     # rojo oscuro para valores altos
],
    zoom=5,
    center=dict(lat=40.0, lon=-3.5),
    mapbox_style="open-street-map",  # <- etiquetas en español
    title="Goles de Messi en La Liga por ciudad 🐐",
    height=650,
)

fig.update_traces(
    textposition="top center",
    textfont=dict(size=11, color="#1A1A2E", family="Arial Black"),
    marker=dict(
        opacity=0.85,
        line=dict(width=2, color="darkred")  # contorno rojo oscuro
    )
)

fig.update_layout(
    title={"x": 0.5},
    coloraxis_colorbar=dict(title="Goles"),
    margin=dict(l=0, r=0, t=50, b=0),
)

fig.show()
```
